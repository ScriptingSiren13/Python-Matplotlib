```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<h2 style="color:#DE006F">CREATING BASIC HISTOGRRAM</h2>

```python
data=[2,5,5,5,7,3,8,7,4]
plt.hist(data)
plt.show()
```

<h2 style="color:#DE006F">CHANGING THE AESTHETICS</h2>

```python
data=[2,5,5,5,7,3,8,7,4]

plt.hist(data,color="#ffb3cb",bins=9)

plt.show()
```

<h2 style="color:#DE006F">WORKING WITH DATASET</h2>

```python
iris=pd.read_csv('iris.csv')
iris.head(2)
```

```python
plt.hist(iris['petal_width'],bins=20,color='#ffb3cb')
plt.show()
```

<h2 style="color:#DE006F">CUMULATIVE HISTOGRAM</h2>

```python
ages=np.random.normal(20,1.5,50)
plt.hist(ages,color="#ffb3cb",cumulative=True,bins=50)
plt.show()
```

```python
ages = np.random.rand(20, 100)  # Generates a 2D array
ages_flattened = ages.ravel()   # Flattens the 2D array into 1D
plt.hist(ages_flattened, color="#ffb3cb", cumulative=True, bins=50)
plt.show()
```
