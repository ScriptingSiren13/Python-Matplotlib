```python

import matplotlib.pyplot as plt
```

```python
langs=["Python","C++","Java","C#","Go"]
votes=[50,24,14,6,17]
explodes=[0,0,0,0.2,0]

plt.pie(votes,labels=langs,explode=explodes,autopct="%.2f%%",pctdistance=1.5,startangle=90)
plt.show()
```

#### LEGENDS

```python

Votes=[10,4,5,17,23]
people=["A","B","C","D","E"]

plt.pie(votes,labels=None)
plt.legend(labels=people)
plt.show()
```

#### PLOT STYLING

```python

from matplotlib import style
#style.use("ggplot")       #you can customize your own stylesheets
style.use("dark_background")
numbers=[10,34,5,33,9]
girls=["Savana","Aliza","Maria","Elle","Prada"]
plt.pie(numbers,labels=None)
plt.legend(labels=girls)
plt.show
```
