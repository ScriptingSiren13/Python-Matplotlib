```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,20,25,30,40]
plt.plot(x,y)

plt.xticks([1,2,3,4,5],["one","two","three","four","five"])
plt.yticks([10,20,25,30,40],["very_low","low","medium","high","very_high"])
plt.show()
```

#### Legends and multiple plots

```python
stock_a=[100,102,99,101,100,102]
stock_b=[90,95,102,104,105,103,109]
stock_c=[110,115,100,105,100,98,95]
plt.plot(stock_a,label="Company1")
plt.plot(stock_b,label="Company2")
plt.plot(stock_c,label="Company3")

#plt.legend() #the legend will be aat the top right
plt.legend(loc="lower center")
```
