```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

<h2 style="color:#FF00FF">Adding titles and labels:</h2>

```python
plt.title("1st line plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
plt.show()
```

<h2  style="color:#FF00FF">CHANGING LINE AESTHETICS:</h2>

```python
plt.plot(x,y,color="pink",linewidth=3,linestyle="-",marker="o")
plt.show()
```

<h2 style="color:#FF00FF">ADDING TWO LINES IN THE SAME PLOT</h2>

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1,color="#DE006F",linestyle=":",linewidth=5)
plt.plot(x,y2,color="purple",linestyle="-",linewidth=5)
plt.title("Line plot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

<h2  style="color:#FF00FF">ADDING SUBPLOT (along the rows)</h2>

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(1,2,1)
plt.plot(x,y1,color="#DE006F",linewidth=2)
plt.title("1st subplot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(color="pink")

plt.subplot(1,2,2)
plt.plot(x,y2,color="purple",linewidth=2)

plt.title("2nd subplot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(color="pink")

plt.tight_layout()  # Adjusts spacing between and around subplots
plt.show()
```

<h2  style="color:#FF00FF">ADDING SUBPLOT (along the columns)</h3></h3>

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(2,1,1)
plt.plot(x,y1,color="#DE006F",linewidth=2)
plt.title("1st subplot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(color="pink")

plt.subplot(2,1, 2)
plt.plot(x,y2,color="purple",linewidth=2)

plt.title("2nd subplot")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(color="pink")

plt.tight_layout()  # Adjusts spacing between and around subplots
plt.show()
```

<h2 style="color:#FF00FF">ADDING X & Y TICKS</h2>

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, linestyle="-", marker="o", color="purple")
plt.xlabel("Prime Numbers")
plt.ylabel("Natural Numbers")
plt.title("Line Plot")
plt.grid(color="pink")

# Customize ticks
plt.xticks([1, 2, 3, 4, 5])  # Specify x-axis tick positions
plt.yticks([0, 2, 4, 6, 8, 12])  # Specify y-axis tick positions

plt.show()
```
