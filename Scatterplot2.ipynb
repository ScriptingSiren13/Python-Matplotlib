```python
import numpy as np
import matplotlib.pyplot as plt
```

<H2 style="color:#FF0D86">CREATING A BASIC SCATTERPLOT</H2>

```python
x=[1,2,3,4,5,6]
y=[10,60,30,80,50,20]

plt.scatter(x,y)
plt.show()
```

<h2 style="color:#FF0D86">CHANGING THE AESTHETICS</h2>

x=[1,2,3,4,5,6]
y=[10,60,30,80,50,20]

plt.scatter(x,y,marker="*",c="#FF0D86",s=200,alpha=0.3)
plt.title("scatterplot")
plt.xlabel('X-axis')
plt.ylabel('Y_axis')
plt.show()

<h2 style="color:#FF0D86">ADDING TWO MARKERS IN THE SAME PLOT</H2>

```python
x=[10,20,30,40,50]
y1=[7,8,5,9,1]
y2=[9,6,8,3,2]
plt.scatter(x,y1,c="#B00058",s=300,marker=".")
plt.scatter(x,y2,c="#FF0D86",s=500,marker="*")

plt.title('scatterplot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

<h2 style="color:#FF0D86">ADDING SUBPLOTS( along the rows)</h2>

```python
x=[10,20,30,40,50]
y1=[7,8,5,9,1]
y2=[9,6,8,3,2]

plt.subplot(1,2,1)
plt.scatter(x,y1,c="#B00058",s=300,marker=".")
plt.title('scatterplot 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')


plt.subplot(1,2,2)
plt.scatter(x,y2,c="#FF0D86",s=500,marker=".")

plt.title('scatterplot 2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.tight_layout()
plt.show()
```

<h2 style="color:#FF0D86">ADDING SUBPLOTS( along the columns)</h2>

```python
x=[10,20,30,40,50]
y1=[7,8,5,9,1]
y2=[9,6,8,3,2]

plt.subplot(2,1,1)
plt.scatter(x,y1,c="#B00058",s=300,marker=".")
plt.title('scatterplot 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')


plt.subplot(2,1,2)
plt.scatter(x,y2,c="#FF0D86",s=500,marker=".")

plt.title('scatterplot 2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.tight_layout()
plt.show()
```

<h2 style="color:#FF0D86">SCATTERPLOT WITH ANNOTATIONS</h2>

```python
x=[1,2,3,4,5,6]
y=[10,60,30,80,50,20]

plt.scatter(x,y,marker="*",c="#FF0D86",s=200,alpha=0.3)

plt.annotate('Point of interest',
            xy=(5,50),
             xytext=(6,80),
            arrowprops=dict(facecolor="#820041",shrink=0.05))
             
plt.title("scatterplot")
plt.xlabel('X-axis')
plt.ylabel('Y_axis')
plt.show()
```
