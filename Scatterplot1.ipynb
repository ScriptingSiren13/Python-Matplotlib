```python
x_data=np.random.random(50)*100
y_data=np.random.random(50)*100
```

#### TO CREATE A SCATTER PLOT

```python
plt.scatter(x_data,y_data)
```

#### TO HAVE A DIFFERENT COLOR: (it can be  a hex-code or rgba as well)

```python

plt.scatter(x_data,y_data,c="pink")
```

#### TO HAVE A DIFFERENT MARKER

```python
plt.scatter(x_data,y_data,marker="*",c="#FF1D8E")
```

#### TO HAVE DIFFERENT SIZES FOR DATA POINTS 

```python
plt.scatter(x_data,y_data,c="#FF1D8E",s=90,marker="*")
```

#### TO AVOID OVERLAP

```python
""""SAY WE HAVE MANY DATA POINTS AND THEY ARE OVERLAPPING: WE use "alpha" to make it transparent. Alpha controls 
the transparency of the points. It is a value between 0 and 1, """
X_data=np.random.random(1000)*100
Y_data=np.random.random(1000)*100
plt.scatter(X_data, Y_data,c="purple",s=90,alpha=0.3)
plt.title("Random Scatter PLot")
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.show
```

### SCATTERPLOT WITH ANNOTATIONS

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
X_data = np.random.random(1000) * 100
Y_data = np.random.random(1000) * 100

# Create a scatter plot for the data
plt.scatter(X_data, Y_data, color='blue', alpha=0.5)

# Annotate a specific point
plt.annotate('Point of Interest', 
             xy=(X_data[50], Y_data[50]), 
             xytext=(X_data[50] + 10, Y_data[50] + 10),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Show the plot
plt.title("Annotated Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.tight_layout()
plt.show()
```
