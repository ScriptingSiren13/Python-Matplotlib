```python
import matplotlib.pyplot as plt
```

```python

fruits=['Apples','Mango','Banana','Raspberry']
quantity=[40,45,24,98]

plt.pie(quantity,
        labels=fruits,
        autopct='%0.1f%%',
        colors=['#FFB3CB','#55002A','#FF0D86','#FFB5EE'],
        radius=2)

plt.pie([1],colors='white',radius=1)
plt.legend(labels=fruits)
plt.show()
```
