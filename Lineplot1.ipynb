```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, linestyle="-", marker="o", color="pink")
plt.xlabel("Prime Numbers")
plt.ylabel("Natural Numbers")
plt.title("Line Plot")
plt.grid(color="red")

# Customize ticks
plt.xticks([1, 2, 3, 4, 5])  # Specify x-axis tick positions
plt.yticks([0, 2, 4, 6, 8, 12])  # Specify y-axis tick positions

plt.show()
```
