```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2 style="color:#FF00BF">VERTICAL BARPLOTS</h2>

```python
students={"Isa":100,"Catherine":89,"Gisele":76}
names=list(students.keys())
values=list(students.values())
plt.bar(names,values)
plt.show()
```

<h2 style="color:#FF00BF">ADDING THE AESTHETICS</h2>

```python
students={"Isa":100,"Catherine":89,"Gisele":76}
names=list(students.keys())
values=list(students.values())
plt.bar(names,values,color="pink",edgecolor="#F9629F",width=1,linewidth=1)
plt.title('MARKS OF THE STUDENTS')
plt.xlabel("Students")
plt.ylabel('Marks')
plt.show()
```

<h2 style="color:#FF00BF">HORIZONTAL BARS</h2>

```python
students={"Isa":100,"Catherine":89,"Gisele":76}
names=list(students.keys())
values=list(students.values())
plt.barh(names,values,color="pink",edgecolor="#F9629F",height=1,linewidth=1)
plt.title('MARKS OF THE STUDENTS')
plt.ylabel("Students")
plt.xlabel('Marks')


plt.show()
```
