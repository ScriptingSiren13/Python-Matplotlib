```python
import pandas as pd
import matplotlib.pyplot as plt
```

<h2 style="color:#DE006F">CREATING ONE BOXPLOT</h2>

```python
one=[1,2,3,4,5,6,7,8,9]

plt.boxplot(one)
plt.show()
```

<H2 style="color:#DE006F">CREATING MULTIPLE BOXPLOTS</H2>

```python
one=[1,2,3,4,5,6,7,8,9]
two=[3,7,6,2,3,9,4,1,2]
three=[9,4,6,4,7,1,3,8,6]

plt.boxplot([one,two,three])
plt.show()
```
