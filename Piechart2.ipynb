<h1 style="color:#DE006F">CREATING A SIMPLE PIE-CHART</h2>

```python
fruits=['Apples','Mango','Banana','Raspberry']
quantity=[40,45,24,98]
plt.pie(quantity,labels=fruits)
plt.show()
```

<H1 style="color:#DE006F">CHANGING THE AESTHETICS</H2>

```python
from matplotlib import style

fruits=['Apples','Mango','Banana','Raspberry']
quantity=[40,45,24,98]

style.use('classic')

plt.pie(quantity,
        labels=fruits,
        autopct='%0.1f%%',
        colors=['#FFB3CB','#55002A','#FF0D86','#FFB5EE'])

plt.legend(labels=fruits)
plt.tight_layout()
plt.show()
```

<h1 style="color:#DE006F">ADDING SOME MORE ATTRIBUTES</h1>

```python

fruits=['Apples','Mango','Banana','Raspberry']
quantity=[40,45,24,98]
explodes=[0,0.1,0,0]

style.use('classic')

plt.pie(quantity,
        labels=fruits,
        autopct='%0.1f%%',
        colors=['#FFB3CB','#55002A','#FF0D86','#FFB5EE'],
        explode=explodes,
        pctdistance=0.8,
        startangle=90)

plt.legend(labels=fruits)
plt.tight_layout()
plt.show()
```
