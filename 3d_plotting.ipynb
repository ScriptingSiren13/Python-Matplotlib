```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

<h1 style="color:#DE006F">EXAMPLE 1</h1>

```python
ax=plt.axes(projection="3d")
x=np.random.random(100)
y=np.random.random(100)
z=np.random.random(100)
ax.scatter(x,y,z)
ax.set_title("3D plot")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.set_zlabel("z-axis")
```

<h1  style="color:#DE006F">EXAMPLE 2</h1>

```python
ax1=plt.axes(projection="3d")

a=np.arange(0,50,0.1)
b=np.sin(a)
c=np.cos(a)
ax1.plot(a,b,c)
ax1.set_title("3D plot")
plt.show()
```

<h1  style="color:#DE006F">SURFACE PLOTS</h1>

```python
 #MESH GRID
ax2=plt.axes(projection="3d")
q=np.arange(-5,5,0.1)
w=np.arange(-5,5,0.1)

X,Y=np.meshgrid(q,w)
Z=np.sin(X)*np.cos(Y)
ax2.plot_surface(X,Y,Z,cmap="magma")
ax2.set_title("3D PLOT")

plt.show()
```

<H1  style="color:#DE006F">SURFACE PLOT- for default top view and Looking along the x-axis</H1>

```python
 #MESH GRID
ax2=plt.axes(projection="3d")
q=np.arange(-5,5,0.1)
w=np.arange(-5,5,0.1)

X,Y=np.meshgrid(q,w)
Z=np.sin(X)*np.cos(Y)
ax2.plot_surface(X,Y,Z,cmap="Purples")
ax2.set_title("3D PLOT")
ax2.view_init(azim=0,elev=90) #for default top view
plt.show()
```
